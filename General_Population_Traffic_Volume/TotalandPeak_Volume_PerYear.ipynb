```python
import os
import glob
import pandas as pd
```

```python
#Reads in the desired year's csv, only choosing coloumns which show location, time, the peak hours and the total volume per day
year = '2020'
filename = year+"_final.csv"
col_list = ['NB_SCATS_SITE','QT_INTERVAL_COUNT','V32','V33','V34','V35','V60','V61','V62','V63','V64','V65','V66','V67','V68','V69','V70','V71','V72','V73','QT_VOLUME_24HOUR']
os.chdir('Sampled_Data_Per_Year/')
df = pd.read_csv(filename,usecols=col_list)
os.chdir('..')
df
```

```python
#Changes the date into just the year and adds up the coloumns from V32 to V35 for the morning peak hour, and the same for evening
df['QT_INTERVAL_COUNT'] = pd.to_datetime(df['QT_INTERVAL_COUNT'])  
df['QT_INTERVAL_COUNT']=df['QT_INTERVAL_COUNT'].dt.year
df['from8to9'] = df.iloc[:,2:7].sum(axis=1)
df['from3to6_30'] = df.iloc[:,7:19].sum(axis=1)
```

```python
df
```

```python
# adds up all the rows which share the same location for: total 24 hour volume, 8 to 9 am peak hour volume and 3 to 6:30 pm 
df2 = df.groupby(['NB_SCATS_SITE','QT_INTERVAL_COUNT'])['QT_VOLUME_24HOUR'].agg(sum).reset_index()
df3  = df.groupby(['NB_SCATS_SITE'])['from8to9'].agg([('from8to9',lambda x: x[x>0].sum())]).reset_index()
df4 = df.groupby(['NB_SCATS_SITE'])['from3to6_30'].agg([('from3to6_30',lambda x: x[x>0].sum())]).reset_index()
df5 = pd.DataFrame({'NB_SCATS_SITE':df4['NB_SCATS_SITE'],'QT_INTERVAL_COUNT':df2['QT_INTERVAL_COUNT'],'from8to9':df3['from8to9'],'from3to6_30':df4['from3to6_30'],'QT_VOLUME_24HOUR':df2['QT_VOLUME_24HOUR']})
df5
```

```python
# Averages the total volumes for average per day over the 90 sampled days and calculates the combined volume during peak hours
df5['from8to9'] = df5['from8to9']/90
df5['QT_VOLUME_24HOUR'] = df5['QT_VOLUME_24HOUR']/90
df5['from3to6_30'] = df5['from3to6_30']/90
df5['combined_peak'] = df5['from8to9']+df5['from3to6_30']
filtered_df = df5
filtered_df
```

```python
final_name = year+'_filtered.csv'
os.chdir("filtered_data/")
filtered_df.to_csv(final_name, index=False)
os.chdir("..")
```
